## 3. Сколько заказов в месяц в среднем не доставляется по разным причинам (вывести детализацию по причинам)?

Первое, что нужно сделать — это исследовать отсутствие доставки и понять, когда мы считаем, что доставка не осуществилась.

У нас есть два столбца: фактическое время доставки `order_delivered_customer_date` и расчетное время доставки `order_estimated_delivery_date`. Логично, что те недоставленные заказы, по которым расчетное время доставки еще не наступило, нельзя включать в расчеты недоставки, так как они еще могут быть доставлены.

А что делать с теми заказами, доставку по которым уже просрочили? Необходимо понять, когда просрочку можно считать фатальной — то есть, такой, при которой заказ с высокой долей вероятности уже не будет доставлен. Посмотрим, какие у нас бывают опоздания доставки.

```python
delays = orders[orders.order_delivered_customer_date > orders.order_estimated_delivery_date].copy()
delays.shape[0]
```

Заказов с опозданием **7827**. Добавим столбец с опозданиями в днях и изучим их:

```python
delays['dls'] = (delays.order_delivered_customer_date - delays.order_estimated_delivery_date).dt.days
delays.head()
```

```python
# строим распределение опозданий
ax = sns.histplot(delays.dls, bins=20)
ax.set(xlabel='Days')
smth = ax.bar_label(ax.containers[0])
```

```python
delays.dls.quantile(0.95)
```

Как видно, **95%** всех опозданий укладывается в **29** дней. Возьмем порог в **30** дней. Таким образом, вероятность,
что заказ доставится, если уже прошло более **30** дней после расчетного времени доставки, крайне мала (меньше **5%**).

Вывод: из недоставленных надо исключить все недавние заказы, для которых опоздание еще не превысило **30** дней, так как возможно, что такие заказы еще будут доставлены.

Введем понятие "сейчас" (now). Так как датафрейм старый, то будем считать, что "сейчас" — это следующий день после максимальной даты.

```python
now = orders[['order_purchase_timestamp',
       'order_approved_at', 'order_delivered_carrier_date',
       'order_delivered_customer_date']].max().max() + timedelta(days=1)
now
```

```python
# для каждого недоставленного заказа посчитаем, сколько дней прошло расчетного времени доставки
orders['delta'] = (now - orders.order_estimated_delivery_date).dt.days
orders.head(2)
```

```python
# создадим таблицу недоставленных заказов, учтя порог в 30 дней
undelivered = orders.sort_values('order_purchase_timestamp').query("order_status != 'delivered' and delta > 30")
undelivered.head(2)
```

Проанализировав статусы, можно выделить следующие причины недоставки: 
* не выставлен счет, 
* нет оплаты, не приступили к сборке, 
* проблема со сборкой заказа, 
* логистическая служба не доставила до покупателя, 
* заказ недоступен,
* заказ отменен.

Созданим отдельную колонку со статусами для удобства дальнейшего анализа

```python
undelivered['non_delivery_reason'] = ''
```

```python
undelivered.loc[orders.order_status == 'created', 'non_delivery_reason'] = 'no invoice'
undelivered.loc[orders.order_status == 'invoiced', 'non_delivery_reason'] = 'no payment'
undelivered.loc[orders.order_status == 'approved', 'non_delivery_reason'] = 'not processed'
undelivered.loc[orders.order_status == 'processing', 'non_delivery_reason'] = 'problems with processing'
undelivered.loc[orders.order_status == 'shipped', 'non_delivery_reason'] = 'no delivery from carrier'
undelivered.loc[orders.order_status == 'unavailable', 'non_delivery_reason'] = 'order is unavailable'
undelivered.loc[orders.order_status == 'canceled', 'non_delivery_reason'] = 'order is canceled'
```

```python
undelivered.head(3)
```

Причины определили. Теперь надо ввести месяцы, чтобы по ним потом группировать. За основу возьмем `order_estimated_delivery_date`, то есть время, когда должны были доставить заказ.

```python
undelivered['order_delivery_month'] = undelivered.order_estimated_delivery_date.dt.strftime('%Y-%m')
undelivered.head(2)
```

Далее группирую по месяцам и причинам недоставки и строю график:

```python
undelivered_orders_by_month = undelivered. \
    query("order_status != 'delivered'"). \
    groupby(['order_delivery_month', 'non_delivery_reason']).size(). \
    to_frame(name='number_of_undelivered_orders').reset_index()

undelivered_orders_by_month.head()
```

```python
plt.figure(figsize=((18, 10)))
ax = sns.barplot(data=undelivered_orders_by_month, 
                 x='order_delivery_month', 
                 y='number_of_undelivered_orders', 
                 hue='non_delivery_reason',
                 palette='bright')
ax.legend(loc='upper left')
plt.setp(ax.get_xticklabels(), rotation=25, ha='right')
ax.set(title='Недоставленные заказы', xlabel='Месяц предполагаемой доставки', ylabel='Количество недоставленных заказов')
sns.despine()
```

По графику уже видно, что лидирует отсутствие доставки от логистов.

Посмотрим на среднемесячную недоставку в разбивке по причинам:

```python
average_undelivered_orders_by_month = undelivered_orders_by_month. \
    groupby('non_delivery_reason'). \
    number_of_undelivered_orders. \
    mean().round(2). \
    to_frame(name='average_undelivered_orders').reset_index(). \
    sort_values('average_undelivered_orders', ascending=False)

average_undelivered_orders_by_month
```

```python
set_style(default_rc=True)
plt.figure(figsize=(7, 4))
ax = sns.barplot(data=average_undelivered_orders_by_month, 
                 x='non_delivery_reason', 
                 y='average_undelivered_orders', 
                 palette='Set1', saturation=0.55)
ax.set(xlabel='Причины недоставки', ylabel='Среднемесячная недоставка', title='Распределение недоставки по причинам')
smth = ax.bar_label(ax.containers[0])
plt.setp(ax.get_xticklabels(), rotation=15, ha='right')
plt.grid(visible=False)
sns.despine(left=True, bottom=True)
```

#### Как видно из графика, основные причины недоставки: 
* проблемы на стороне логистики, 
* заказ недоступен,
* заказ отменен.

## 4. По каждому товару определить, в какой день недели товар чаще всего покупается.

Создадим новую таблицу `orders_with_products`, добавив в `orders` информацию о товарах из `products` и удалив ненужные столбцы.

```python
orders_with_products = orders.merge(products[['order_id', 'order_item_id', 'product_id']], on='order_id')
```

```python
# конечно, выделяем только успешные заказы
orders_with_products = orders_with_products.query("order_status == 'delivered'")
```

```python
orders_with_products = orders_with_products.loc[:, ['order_id', 
       'customer_id', 'customer_unique_id', 'order_item_id', 'product_id',
       'order_purchase_timestamp']]
orders_with_products.head()
```

Введем еще одну колонку с днями недели:

```python
orders_with_products['order_purchase_weekday'] = orders_with_products.order_purchase_timestamp.dt.day_name()
orders_with_products.head(3)
```

Теперь посчитаем, в каком количестве тот или иной товар был куплен в какой день недели за все время и выведем для каждого товара лучший день недели. Отсюда можно понять, в какой день недели товар лучше продавался.

```python
products_best_weekday = orders_with_products. \
groupby(['product_id','order_purchase_weekday'],as_index=False). \
order_id.count().sort_values('order_id', ascending=False). \
rename(columns={'order_id': 'number_of_products', 'order_purchase_weekday': 'best_weekday'}). \
groupby('product_id').head(1)

products_best_weekday.head()
```

Вывели лучший день недели для каждого товара. Но здесь есть нюанс. Посмотрим, сколько было продаж у каждого товара за все время.

```python
number_of_sold_products = orders_with_products. \
product_id.value_counts().to_frame(name='number_of_products').reset_index(). \
rename(columns={'index': 'product_id'})
number_of_sold_products.tail()
```

```python
number_of_sold_products.number_of_products.quantile(0.75)
```

Как видим, 75% всех товаров было продано в количестве не более 3-х штук каждого товара. Для таких товаров поиск лучшего дня недели не имеет смысла. Возможно, имеет смысл ограничить наш результат товарами, которых продали от 10 шт.

```python
ids = number_of_sold_products.query("number_of_products >= 10").product_id
```

```python
products_best_weekday.query("product_id in @ids")
```

Таких товаров 1884. Но есть еще один нюанс. Как можно предположить, для некоторых товаров может обнаружится несколько дней недели с максимальным количеством продаж. Поэтому, лучше вывести для каждого товара все лучшие дни недели.

```python
# выведим для каждого товара и каждого дня недели количество проданных товаров
result = orders_with_products. \
groupby(['product_id','order_purchase_weekday'],as_index=False). \
order_id.count().query("product_id in @ids"). \
rename(columns={'order_id': 'number_of_products'})
result
```

```python
# смерджим с "product_best_weekday", чтобы добавить максимальное количество товара, проданного в день недели
result = result.merge(products_best_weekday[['product_id', 'number_of_products']], on='product_id')
result.head(3)
```

```python
# подправим название столбцов
result.columns = ['product_id', 'best_weekday', 'number_of_products',
       'max_number_of_products']
result.head(3)
```

```python
# отфильтруем те строчки, где количество проданных товаров совпадает с максимальным
result = result[result.number_of_products == result.max_number_of_products][['product_id', 'best_weekday', 'number_of_products']]
result.head(3)
```

```python
result = result. \
groupby(['product_id', 'number_of_products'], as_index=False). \
agg({'best_weekday': pd.Series.to_list})[['product_id', 'best_weekday', 'number_of_products']]. \
sort_values('number_of_products', ascending=False)
result['best_weekday'] = result.best_weekday.apply(lambda x: ', '.join(x))
result
```

Итого получили таблицу с лушим днем/днями недели для каждого товара, которого было продано от 10 шт.

## 5. Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)?

Посчитаем, сколько у каждого пользователя в каждый месяц было покупок, предварительно введя колонку с месяцем покупки

```python
orders_with_products['order_purchase_month'] = orders_with_products.order_purchase_timestamp.dt.strftime('%Y-%m')
orders_with_products.head(2)
```

```python
purchases_by_customer_by_month = orders_with_products. \
    groupby(['customer_unique_id', 'order_purchase_month']). \
    order_id.count(). \
    to_frame(name='purchases').reset_index()

purchases_by_customer_by_month.head()
```

Учтем тот факт, что внутри месяца может быть не целое количество недель. Посчитаем количество недель для каждого месяца:

```python
purchases_by_customer_by_month['num_of_weeks_in_month'] = (pd.
                                        to_datetime(purchases_by_customer_by_month.order_purchase_month).
                                        dt.days_in_month) / 7
purchases_by_customer_by_month.head(3)      
```

Посчитаем среднее количество покупок в неделю у каждого покупателя по каждому месяцу:

```python
purchases_by_customer_by_month['avg_purchases_by_week'] = (purchases_by_customer_by_month.purchases /
                                                       purchases_by_customer_by_month.num_of_weeks_in_month)
purchases_by_customer_by_month.head()
```

Представим данные в удобном виде — в виде сводной таблицы:

```python
avg_weekly_purchases_by_month = purchases_by_customer_by_month.pivot(index='customer_unique_id', 
                                     columns='order_purchase_month', 
                                     values='avg_purchases_by_week').fillna(0).reset_index().round(3)
avg_weekly_purchases_by_month.rename_axis(None, axis=1, inplace=True)
avg_weekly_purchases_by_month.head()
```

**Получили таблицу, где для каждого пользователя по каждому месяцу посчитано средненедельное количество покупок.**

Но, учитывая, что почти все наши клиенты совершили 1-2 покупки за ВСЕ время, то такая таблица едва ли имеет смысл...

## 6. Когортный анализ

Для начала создадим отдельную таблицу для удобства, оставив нужные столбцы. И добавим столбец с месяцами покупки.

```python
cohort_analysis = (orders.query("order_status == 'delivered'").
 loc[:, ['order_id', 'customer_unique_id', 'order_purchase_timestamp']]
)
cohort_analysis.head(3)
```

```python
cohort_analysis['order_purchase_month'] = cohort_analysis.order_purchase_timestamp.dt.strftime('%Y-%m')
cohort_analysis.head(2)
```

В наших данных полный год со всеми месяцами — только **2017-й**. Поэтому для когортного анализа возьмем пользователей, которые совершили первую покупку в 2017-м году.

Сначала посчитаем месяц первой покупки для каждого юзера:

```python
first_purchase_month = cohort_analysis. \
                       groupby('customer_unique_id'). \
                       order_purchase_month.min(). \
                       to_frame(name='first_purchase_month'). \
                       reset_index()
first_purchase_month.head(3)
```

Вставляем посчитанные месяцы в наш датафрейм `cohort_analysis`:

```python
cohort_analysis = cohort_analysis.merge(first_purchase_month, on='customer_unique_id')
cohort_analysis.head(3)
```

```python
users_from_2017 = (cohort_analysis[(cohort_analysis.first_purchase_month >= '2017-01') & 
                   (cohort_analysis.first_purchase_month <= '2017-12')].customer_unique_id
                  )
```

```python
# отбираем тех, кто совершил первую покупку в 2017-м
cohort_analysis = cohort_analysis.query('customer_unique_id in @users_from_2017')
```

Теперь необходимо посчитать для каждого месяца покупки каждого клиента количество месяцев, прошедшие с первой покупки. Так мы поймем, через сколько месяцев после первой покупки клиент совершил следующую и так далее покупки. Это поможет рассчитать **retention**.

Создадим для этого вспомогательный датафрейм `cohort_period_df`:

```python
# добавим уникальные id - потом понадобятся для корректного мёрджа
cohort_analysis['id'] = range(cohort_analysis.shape[0])
```

```python
cohort_period_df = cohort_analysis.loc[:, ['id', 'order_purchase_month', 'first_purchase_month']].copy()
cohort_period_df[['order_purchase_month', 
               'first_purchase_month']] = cohort_period_df[['order_purchase_month', 
                                                         'first_purchase_month']].apply(pd.to_datetime)
cohort_period_df.head()
```

```python
cohort_period_df['cohort_period'] = (cohort_period_df.
                                     order_purchase_month.
                                     dt.to_period('M').
                                     astype(int) - cohort_period_df.
                                     first_purchase_month.dt.to_period('M').astype(int)
                                    )
cohort_period_df.head()
```

Вставим полученный столбец `cohort_period` в датафрейм `cohort_analysis`:

```python
cohort_analysis = cohort_analysis.merge(cohort_period_df[['id', 'cohort_period']], on='id')
del cohort_analysis['id']
cohort_analysis.head()
```

Посмотрим, какие у нас образовались периоды:

```python
np.sort(cohort_analysis.cohort_period.unique())
```

Представлены все периоды c 0 по 19, кроме 18. Добавим искусственно **18-й**:

```python
cohort_analysis.loc[len(cohort_analysis.index)] = ['', '', '', '', '2018-01', 18]
cohort_analysis.tail()
```

Строим сводную таблицу:

```python
user_retention = cohort_analysis.pivot_table(values='customer_unique_id', 
                                 index='first_purchase_month', 
                                 columns='cohort_period', 
                                 aggfunc=pd.Series.nunique,)
user_retention.drop(['2018-01'], inplace=True) # удаляем искусственную строку
user_retention
```

В нашей сводной таблице `user_retention` пока представлено количество заказов в абсолютных величиных. А нам надо в процентах от первоначальных цифр, чтобы посчитать **retention rate**

```python
user_retention_prcnt = user_retention.divide(user_retention[0], axis=0)

# удаляем столбец с первой покупкой - там и так понятно, что 100%, и это бы сильно искажало цветовую шкалу 
del user_retention_prcnt[0]
user_retention_prcnt
```

Визуализируем полученные данные для наглядности, чтобы упростить их анализ.

```python
set_style(yticks='on')
plt.figure(figsize=(18, 10))
ax = sns.heatmap(user_retention_prcnt, annot=True, fmt='.2%')
ax.set_yticklabels(ax.get_yticklabels(), rotation=0)
ax.set(title='Cohorts: User Retention', ylabel='Cohorts by First Purchase Month', xlabel='Cohort Period')
sns.despine()
```

Согласно полученному графику `Heatmap`, самый высокий **retention** на **3-й** месяц покупки у когорты **2017-06**. При этом, очень близко к нему оказалась когорта **2017-03**.

Также видно, что retention в целом очень низкий и не превышает **0,72%**. В основном клиенты покупают только **1** раз.

Альтернативная визуализация, где цветовая градация отдельно по каждому периоду:

## RFM-анализ

Создадим датафрейм под RFM анализ:

```python
orders_rfm = orders.query("order_status == 'delivered'").loc[:, ['order_id', 'customer_unique_id',
       'order_purchase_timestamp']]
orders_rfm['order_purchase_month'] = orders_rfm.order_purchase_timestamp.dt.strftime('%Y-%m')
orders_rfm.head(3)
```

Добавим в таблицу `orders_rfm` сумму заказа:

```python
order_amount = products.groupby('order_id').price.sum().sort_values().to_frame(name='order_amount').reset_index()
orders_rfm = orders_rfm.merge(order_amount, on='order_id')
orders_rfm.head(3)
```

Посчитаем, сколько дней прошло с последней покупки (**now** уже было посчитано ранее).

```python
orders_rfm['since_last_order'] = orders_rfm['order_purchase_timestamp'].apply(lambda x: (now - x).days)
orders_rfm.head(3)
```

Посчитаем RFM и создадим новую таблицу с результатами расчета

```python
rfm = orders_rfm. \
groupby('customer_unique_id', as_index=False). \
agg({'since_last_order': 'min', 'order_id': 'count', 'order_amount': sum}). \
rename(columns={'since_last_order': 'Recency', 'order_id': 'Frequency', 'order_amount': 'Monetary'})

rfm.head(3)
```

Теперь созданим систему скоринга для каждого параметра.

```python
rfm.Monetary.describe()
```

```python
rfm.Monetary.quantile(0.95)
```

То есть **95%** всех заказов делаются на сумму до **420**

Также, как мы выяснили ранее, **94.2%** пользователей совершили только **1** покупку.

Проиллюстрируем:

```python
sns.histplot(data=rfm, x='Monetary', bins=100)
```

```python
ax = sns.histplot(data=rfm, x='Frequency')
smth = ax.bar_label(ax.containers[0])
```

Для **RFM** анализа необходимо проставить категории для каждого параметра анализа. Обычно рекомендуют делать 3-х или 5-ти бальную шкалу. После некоторых изыскаий я решил поступить следующим образом. Для построения графиков двумерных зависимостей (FM, RM, FR) я создам 5-ти бальную шкалу, так как это даст большую наглядность. А для создания сегментов буду использовать 3-х бальную шкалу, чтобы не плодить огромное количество сегментов. При этом, и в том и в другом случае `Frequency` будет 3-х бальной по той причине, что на **1-2** покупки проходится почти 100% всех данных (видно из графика выше).

Итак, сначала строим 5-бальную шкалу. Чем выше бал, тем выше мы оцениваем клиента по соответствующему параметру.

Но, как видно из графиков выше, данные очень неравномерные, собраны в левой части графиков и с длинными хвостами выбросов. Поэтому делать равномерные категории не имеет смысла.

Пэтому буду использовать такую концепцию: делаем неравномерные категории для `Frequency` и `Monetary`. Для `Monetary`: данные до **95-го** процентиля разбиваем на **4** категории равными диапазонами, а **5-ю** отдаем последним **5%**. Что касается `Frequency`, то ее делим таким образом:
* 1п — 1 категория, 
* 2п — 2 кат.,
* остальные — 3 кат.

`Recency` делим на **5** равномерных диапазона, так как там нет перекоса данных в какую-либо строну.

Создаю категории:

```python
# найдем максимум у Recency, чтобы потом разбить на равные промежутки
rfm.Recency.max()
```

```python
rfm['R'] = pd.cut(rfm.Recency, [0, 152, 305, 457, 610, 800], labels=[5, 4, 3, 2, 1])
rfm['F'] = pd.cut(rfm.Frequency, [0, 1, 2, 20], labels=[1, 2, 3])
rfm['M'] = pd.cut(rfm.Monetary, [0, 105, 210, 315, 420, 15000], labels=[1, 2, 3, 4, 5])
```

```python
rfm.head()
```

```python
rfm.R = pd.to_numeric(rfm.R)
rfm.F = pd.to_numeric(rfm.F)
rfm.M = pd.to_numeric(rfm.M)
```

Создам RM, RF и FM сочетания и проиллюстрирую зависимость:

```python
rm = rfm.pivot_table(values='customer_unique_id', index='M', columns='R', aggfunc='count')
rf = rfm.pivot_table(values='customer_unique_id', index='F', columns='R', aggfunc='count')
fm = rfm.pivot_table(values='customer_unique_id', index='M', columns='F', aggfunc='count')

set_style(default_rc=True, xticks='off')
fig, axes = plt.subplots(3, 1, figsize=(5, 9))
tables = [rm, rf, fm]
for table, ax in zip(tables, axes.flatten()):
    sns.heatmap(table, annot=True, fmt="d", ax=ax)
plt.tight_layout()
```

Из графиков видно, что основная масса клиентов - это те, кто совершил покупки сравнительно недавно, причем на небольшую сумму, и для большинства это была их первая покупка. То есть интернет-магазин живет в основном за счет притока большого количества новых клиентов с небольшими суммами разовых покупок. Это довольно шаткая ситуация, так как по сути это путь экстенсивного развития. Что будет, когда поток новых клиентов уменьшится (а он уменьшится ибо количество потенциальных клиентов ограничено)?

Теперь создадим сегменты. Они помогут выделить разные кластеры аудитории, с которым можно работать отдельно, чтобы, например, исправить проблему, описанную выше. Для этого используем уже 3-х бальную шкалу, чтобы было поменьше сегментов, и с ними было проще работать и анализировать.

Создадим новую таблицу для этого:

```python
rfm_seg = rfm.copy()
```

Создадим категории:

```python
rfm_seg['R'] = pd.cut(rfm_seg.Recency, [0, 254, 508, 800], labels=[3, 2, 1])
rfm_seg['F'] = pd.cut(rfm_seg.Frequency, [0, 1, 2, 20], labels=[1, 2, 3])
rfm_seg['M'] = pd.cut(rfm_seg.Monetary, [0, 210, 420, 15000], labels=[1, 2, 3])
```

```python
rfm_seg.head()
```

```python
rfm_seg.R = pd.to_numeric(rfm_seg.R)
rfm_seg.F = pd.to_numeric(rfm_seg.F)
rfm_seg.M = pd.to_numeric(rfm_seg.M)
```

```python
# соединив R, F и M, получим RFM Score
rfm_seg['RFM_Score'] = rfm_seg['R'].map(str) + rfm_seg['F'].map(str) + rfm_seg['M'].map(str)
rfm_seg.head()
```

Теперь у нас есть **27** RFM Scores от **111** (наихудший) до ***333*** (наилучший). На основании них создадим сегменты.

```python
seg_map = {
    r'1[1-2][1-2]': 'Потерянные не важные',
    r'1[1-2]3': 'Потерянные хорошие',
    r'131': 'Потерянные постоянные эконом',
    r'132': 'Потерянные хорошие и постоянные',
    r'133': 'Потерянные VIP',
    r'2[1-2][1-2]': 'Спящие не важные',
    r'2[1-2]3': 'Спящие хорошие',
    r'231': 'Спящие постоянные эконом',
    r'232': 'Спящие хорошие постоянные',
    r'233': 'Спящие VIP',
    r'311': 'Новички эконом',
    r'312': 'Новички средние',
    r'3[1-2]3': 'Потенциальные VIP',
    r'321': 'Потенциально лояльные эконом',
    r'322': 'Потенциально хорошие лояльные',
    r'331': 'Лояльные эконом',
    r'332': 'Лояльные хорошие',
    r'333': 'VIP'
}
```

```python
rfm_seg['Segment'] = rfm_seg.RFM_Score
rfm_seg['Segment'] = rfm_seg.Segment.replace(seg_map, regex=True)
rfm_seg.head()
```

Посмотрим, как у нас распределились клиенты по сегментам.

```python
for_plot = rfm_seg.Segment.value_counts().to_frame(name='number_of_users').reset_index()
labels = ['{:} ({:.2g}%)'.format(int(value), 
                                 value*100/for_plot.number_of_users.sum()) for value in for_plot.number_of_users]
set_style(xticks='off')
plt.figure(figsize=(18, 10))
ax = sns.barplot(data=for_plot, x='number_of_users', y='index', palette='muted')
plt.grid(False)
ax.set(xlabel='', ylabel='Сегменты', xticklabels='', title='Распределение клиентов по сегментам')
ax.bar_label(ax.containers[0], labels=labels, padding=10)
sns.despine(left=True, bottom=True)
```

Как мы видим, здесь явная проблема с лояльными постоянными клиентам. Основная масса (89%): экономные новички, а также спящие либо вообще потерянные клиенты с небольшыми потраченными суммами. Можно попробовать вернуть часть спящих клиентов, а может даже реанимировать потерянных. Это должна быть не слишком дорогая массовая стимуляция повторных продаж и увеличения среднего чека. Есть также хорошая новость, что у нас относительно немалое количество потенциальных VIP (2180), а также средних новичков (4032) и хороших, но спящих (1828). На них стоит сосредоточить усилия и перевести их в разряд VIP и хороших лояльных клиентов. Можно также вернуть спящих VIP, тем более, что шанс хороший: из VIP потерян за все время тольно 1.
